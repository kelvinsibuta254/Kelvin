```python
import pandas as pd
import numpy as np
```

```python
#Working with Dictionaries
data = {
    "Name": ["Kelvin", "Simiyu", "Sibuta", "Kayra"],
    "Age" : [25, 25, 27, 2]
}
print(type(data))
# Covert a dictionary into a DataFrame, assign the result to the variable name df
df = pd.DataFrame(data)
df
```

```python
#Set index
df.set_index("Name", inplace  = True)
print(df)
```
